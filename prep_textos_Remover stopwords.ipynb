```
sentenca = "O IGTI oferece especialização em Deep Learning. Deep Learning e utilziado em diversas aplicacoes. As aplicacoes de deep learning nesta. O IGTI ambém oferece bootcamp"
```

```
sentenlower = sentenca.lower()
```

```
tokenixacao_sentencas1 = nltk.sent_tokenize
sample_sentence3= tokenixacao_sentencas1(text=sentenlower)
pprint(sample_sentence3)
```

```
#selecionar a primeira sentençca
sample_sentence3[0]
```

```
#Tokenizacao das palvras
tokenizacao_palavras2 = nltk.word_tokenize
list_words= []

for i in range(len(sample_sentence3)):
  sample_word = tokenizacao_palavras2(text=sample_sentence3[i])
  list_words.extend(sample_word)
```

```
print(list_words)
```

```
#tokenizacao palavras 
def tokenizaPalvras(sentenca):
  sample_words = tokenizacao_palavras(text=sentenca)
  return sample_words

#Remover stopwords, criando dicionário de stopwords
def removeStopWords (list_of_words):
  my_stopwords = ["o","em","as","de","sao","nesta",".","e","a","na","do",]
  list_cleaned = set(list_of_words)-set(my_stopwords)
  return list_cleaned


my_boW = removeStopWords(list_words)
```

```
print(my_boW)
```

```
#Criar Vetor para aplicar o Bag of Words
def bagofwords (sentence,words):
  sentence_words = tokenizacao_palavras(sentence)
  #conta a sequencia da palavra que está no vetor do BOW
  bag = np.zeros(len(words))
  for sw in sentence_words:
    for i,word in enumerate(words):
      if word == sw:
        bag[i] += 1

  return np.array(bag)
```

```

sentenca_teste = "o igti oferece especializacao em deep learning e o igti oferce boot camp "
print(bagofwords(sentenca_teste,my_boW))
```
